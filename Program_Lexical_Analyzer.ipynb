```python
#lexical analyzer
import string

#input
print('Daftar Kata:\n Apa (Ayah),\n Anya (Ibu),\n Tofu (Tahu),\n Tempeh (Tempe),\n Torta (Kue),\n Asztal (Meja),\n Papucs (Sandal),\n Viselet (Memakai),\n Olvas (Membaca),\n Eszik (Makan)')
kalimat = input('Masukkan kata: ')
kata_input = kalimat.lower().replace(" ", ", ") +' #'

#inisialisasi
list_alfabet = list(string.ascii_lowercase)
list_state = ['q0', 'q1', 'q2', 'q3', 'q4', 'q5', 'q6', 'q7', 'q8', 'q9', 'q10', 'q11', 'q12', 'q13', 'q14', 'q15', 'q16', 'q17', 'q18', 'q19', 'q20', 'q21', 'q22', 'q23', 'q24', 'q25', 'q26', 'q27', 'q28', 'q29', 'q30', 'q31', 'q32', 'q33', 'q4']

tabel_transisi = {}

for state in list_state:
  for alfabet in list_alfabet:
    tabel_transisi[(state, alfabet)] = 'error'
    tabel_transisi[(state, '#')] = 'error'
    tabel_transisi[(state, ' ')] = 'error'

#input diawali spasi
tabel_transisi['q0', ' '] = 'q0'
tabel_transisi[('q1', ',')] = 'q6'
tabel_transisi[('q1', ' ')] = 'q6'
tabel_transisi[('q6', ' ')] = 'q0'

#update the transition table for the following token : Apa
tabel_transisi[('q0', 'a')] = 'q7'
tabel_transisi[('q7', 'p')] = 'q9'
tabel_transisi[('q9', 'a')] = 'q1'

#update the transition table for the following token : Anya
tabel_transisi[('q0', 'a')] = 'q7'
tabel_transisi[('q7', 'n')] = 'q8'
tabel_transisi[('q8', 'y')] = 'q9'
tabel_transisi[('q9', 'a')] = 'q1'

#update the transition table for the following token : Asztal
tabel_transisi[('q0', 'a')] = 'q7'
tabel_transisi[('q7', 's')] = 'q2'
tabel_transisi[('q2', 'z')] = 'q3'
tabel_transisi[('q3', 't')] = 'q4'
tabel_transisi[('q4', 'a')] = 'q5'
tabel_transisi[('q5', 'l')] = 'q1'

#update the transition table for the following token : Tofu
tabel_transisi[('q0', 't')] = 'q10'
tabel_transisi[('q10', 'o')] = 'q11'
tabel_transisi[('q11', 'f')] = 'q12'
tabel_transisi[('q12', 'u')] = 'q1'

#update the transition table for the following token : Tempeh
tabel_transisi[('q0', 't')] = 'q10'
tabel_transisi[('q10', 'e')] = 'q15'
tabel_transisi[('q15', 'm')] = 'q16'
tabel_transisi[('q16', 'p')] = 'q17'
tabel_transisi[('q17', 'e')] = 'q18'
tabel_transisi[('q18', 'h')] = 'q1'

#update the transition table for the following token : Torta
tabel_transisi[('q0', 't')] = 'q10'
tabel_transisi[('q10', 'o')] = 'q11'
tabel_transisi[('q11', 'r')] = 'q13'
tabel_transisi[('q13', 't')] = 'q14'
tabel_transisi[('q14', 'a')] = 'q1'

#update the transition table for the following token : Papucs
tabel_transisi[('q0', 'p')] = 'q27'
tabel_transisi[('q27', 'a')] = 'q28'
tabel_transisi[('q28', 'p')] = 'q29'
tabel_transisi[('q29', 'u')] = 'q30'
tabel_transisi[('q30', 'c')] = 'q31'
tabel_transisi[('q31', 's')] = 'q1'

#update the transition table for the following token : Viselet
tabel_transisi[('q0', 'v')] = 'q32'
tabel_transisi[('q32', 'i')] = 'q33'
tabel_transisi[('q33', 's')] = 'q34'
tabel_transisi[('q34', 'e')] = 'q35'
tabel_transisi[('q35', 'l')] = 'q36'
tabel_transisi[('q36', 'e')] = 'q37'
tabel_transisi[('q37', 't')] = 'q1'

#update the transition table for the following token : Olvas
tabel_transisi[('q0', 'o')] = 'q19'
tabel_transisi[('q19', 'l')] = 'q20'
tabel_transisi[('q20', 'v')] = 'q21'
tabel_transisi[('q21', 'a')] = 'q22'
tabel_transisi[('q22', 's')] =  'q1'

#update the transition table for the following token : Eszik
tabel_transisi[('q0', 'e')] = 'q23'
tabel_transisi[('q23', 's')] = 'q24'
tabel_transisi[('q24', 'z')] = 'q25'
tabel_transisi[('q25', 'i')] = 'q26'
tabel_transisi[('q26', 'k')] = 'q1'

#end transastion
tabel_transisi[('q1', '#')] = 'accept'
tabel_transisi[('q6', '#')] = 'accept'

#lexical analysis
idx_char = 0
state = 'q0'
current_token = ''
while state!='accept':
    current_char = kata_input[idx_char]
    current_token += current_char
    state = tabel_transisi[(state, current_char)]
    # print(current_char)
    # print(state)
    if state == 'q6':
        print('kata:', current_token,'valid')
        current_token = ''
    if state == 'error':
        print('error')
        break
    idx_char = idx_char + 1

#conclusion
if state == 'accept':
	print('semua kata: ', kalimat,', valid') 
```
